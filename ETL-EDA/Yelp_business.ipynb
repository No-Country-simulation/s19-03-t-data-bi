```python
def expand_json_columns(df):
    """
    Expande columnas de un DataFrame que contengan JSON o diccionarios utilizando pd.json_normalize.
    """
    for col in df.columns:
        try:
            # Intenta normalizar la columna
            expanded = pd.json_normalize(df[col].dropna())
            # Renombra las columnas nuevas con el prefijo de la columna original
            #expanded.columns = [f"{col}.{subcol}" for subcol in expanded.columns]
            # AÃ±ade las columnas nuevas al DataFrame
            df = pd.concat([df.drop(columns=[col]), expanded], axis=1)
        except:
            # Si no se puede normalizar, pasa a la siguiente columna
            continue

    return df
```

```python
def clean_and_convert(value):
    if isinstance(value, str):  # Si es una cadena de texto
        try:
            # Eliminar prefijos 'u' no deseados
            value = value.replace(" u'", " '").replace(' u"', ' "')
            return ast.literal_eval(value)
        except (ValueError, SyntaxError):
            return value  # Devolver como estÃ¡ si no es un diccionario vÃ¡lido
    return value
```

```python
col_complejas['BusinessParking'] = col_complejas['BusinessParking'].apply(clean_and_convert)
```
